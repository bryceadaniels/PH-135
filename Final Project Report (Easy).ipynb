### Introduction
   Superconductivity is a highly researched field within physics and material science. Superconductors are materials that have no electrical resistivity when the temperature drops below a material specific critical temperature. The data for this report comes from a dataset posted on Kaggle.com by Kam Ham idieh titled "Superconductivity Data Data Set". This data contains many features, including critical temperature, extracted from over 21,000 superconductors. It also contains the chemical composition of each of the superconductors tested.
   
   There are different types of superconductors. Type I superconductors are elements that naturally superconductive but the critical temperature is extremely low (less than 10 K). Type II superconductors are alloys that have superconductive properties at higher temperatures (80 K). Because of these two types, it is hypothesized that as more elements are combined, the elements can have a higher critical temperature.  

## Methods & Results

To find how the number of elements is related to the critical temperature, the critical temperature of each material tested was plotted based on it's number of elements used (plotted in red). It also was placed into a list with other materials with the same number of elements. These critical temperatures were then averaged and plotted in blue on the graph below. 




As seen by the trend of the blue stars, there is a trend of the critical temperature increasing as the number of elements in the material increases. These points were then curve fitted for both linear and quadratic functions. 
The linear function came to be $T_c = 10.874n-10.618$ where $n$ is the number of elements. The error in this fit is 601.350. When fitted to a quadratic, the function is $T_c = -0.45n^2 + 15.39n-18.90$ with an error of 538.562. 
These functions are shown on the graph below.








The question may be asked, "If the more elements used increases the critical temperature, why do we not make materials of large amounts of elements to have room temperature superconductors?" This question is answered when looking at the atomic masses compared to the critical temperatures. These two features plotted together shows that higher mass does not mean higher critical temperature (See graph below). Instead, it shows that there is a "sweet spot" between 50 and 125 amu. When superconductors have an atomic mass higher than that range, the critical temperature drops significantly. 




## Conclusion

As shown in this data, there is a correlation between the number of elements used and the corresponding critical temperature of a material. This agrees with the differences between Type I and Type II superconductors. However, this report also shows a challenge facing research scientists of being able to create superconductors that have high critical temperatures but are not too atomically heavy. When materials are discovered that have higher critical temperatures, there will be massive benefits to society. This is the challenge the future holds for materials physics.
