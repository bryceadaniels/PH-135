11.  Below you will find 5 equal-length arrays (t,x,y,vx,vy).  Turn them into a pandas dataframe so that each array is its own column.  Set the column labels to be "time", "x", "y", "vx", and "vy".


```python
from numpy import arange,ones_like,pi,cos,sin

vi = 30
theta = 35 * pi /180
a = -9.8
yi = 50

t = arange(0,8,.1)
x = vi * cos(theta) * t 
y = yi + vi * sin(theta) * t + 1/2 * a * t**2
vx = ones_like(t) * vi * cos(theta)
vy = vi * sin(theta) + a * t

df = pd.DataFrame(np.transpose([t,x,y,vx,vy]),columns = ["time","x","y","vx","vy"])
df
```

12. Use `sympy` to solve the following equation for x. $$5 x^2 + 3x - 8 = 2$$


```python
x = sp.Symbol("x")
eq = sp.Eq(5*x**2+3*x-8,2)
sp.solveset(eq,x)
```

13. Use sympy to simplify the following expression. $$x^2 z + 4 x y z + 8 y z$$


```python
x,y,z=sp.symbols("x y z")
eq = x**2 * z + 4*x*y*z+8*y*z
sp.simplify(eq)
```

Plot this function from $-5 < x < 10$, $-3 < y < 8$.   
   Hints:
     1. To increase the size of the plot add the following: `pyplot.figure(figsize = (30,30))`.
     2. Restrict the z range with `ax.set_zlim(0,7.5e-9)` where `ax` is the plot axes.


```python
x = np.linspace(-5,10,200)
y = np.linspace(-3,8,200)
X,Y = np.meshgrid(x,y)
k = 8.99e9
e = 1.602e-19
V = k*e*(1/np.sqrt((x-5)**2+y**2)+1/np.sqrt(x**2 + (y-5)**2)+1/np.sqrt(x**2+y**2))
plt.figure(figsize=(30,30))
ax = plt.axes(projection = '3d')
ax.plot(V,x,y)
```
