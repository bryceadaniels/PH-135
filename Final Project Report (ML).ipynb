### Introduction
   Superconductivity is a highly researched field within physics and material science. Superconductors are materials that have no electrical resistivity when the temperature drops below a material specific critical temperature. The data for this report comes from a dataset posted on Kaggle.com by Kam Ham idieh titled "Superconductivity Data Data Set". This data contains many features, including critical temperature, extracted from over 21,000 superconductors. It also contains the chemical composition of each of the superconductors tested.
   
   There are different types of superconductors. Type I superconductors are elements that naturally superconductive but the critical temperature is extremely low (less than 10 K). Type II superconductors are alloys that have superconductive properties at higher temperatures (80 K). The purpose of this research is to see which of the statistical features of this data set best correlates with the critical temperature. 

### Methods
To find the characteristic that correlates with the critical temperature, the maximal information coefficient (MIC) was calculated for each variable as described by Kevin Murphy. Murphy explains the three equations below. $\mathbb{I}$ is the mutual information which "determine[s] how similar the joint distributuion of $p(X,Y)$ is to the factored distribution $p(X)p(Y)$." The data is divided into bins in the x and y direction to quantize the mutual information. The number of bins directly impacts the $\mathbb{I}$ value. $m(x,y)$, as shown below, takes the maximum value for $\mathbb{I}$ and divides it by the log of the smaller number of bins.  The MIC value is defined as the maximum of that $m(x,y)$ function where $xy$ is less than $B$ and $B$ is defined as $B = N^{0.6}$. 


The figure below shows the code used to calculate the MIC value of the variables. In the first cell on the top left, the function to calculate the mutual information is defined. The cell below and the cell on the right are one continuous cell broken up. It would take a very long time to calculate the MIC for every single possible number of bins, so it was determined to only have 10 MIC values for each variable. The bin sizes for those 10 values were randomized. The right code block is the creation of the grid, counting of the number of data points in each bin, and calculation of the MIC values. There are definitely other ways to do this that are much more efficient, but this was a good first attempt.
