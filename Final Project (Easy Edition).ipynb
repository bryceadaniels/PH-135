```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Finding Highest Critical Temp

```python
data1 = pd.read_csv('SuperConductor-1.csv')
data2 = pd.read_csv('unique_m.csv')
maxTc = data2['critical_temp'].max()
highest_material=data2[data2['critical_temp']==maxTc]['material']
max_material = data2.at[20573,'material']
print(f'The highest critical temp is {maxTc} in the material {max_material}')
data1
```

### Plotting avg critical temp by element

```python
temps = []
for column in data2.columns[:-2]:
    avg_critical_temp= data2[data2[column]>0]['critical_temp'].mean()
    temps.append(avg_critical_temp)
plt.figure(figsize=(30,20))
plt.bar(data2.columns[:-2],temps)
plt.title('Superconductivity with Elements',fontsize=60)
plt.xlabel('Elements',fontsize = 30)
plt.ylabel('Average Critical Temperature (K)',fontsize = 30)
```

### Plotting number of elements with critical temp

```python
cTemps1 = []
cTemps2 = []
cTemps3 = []
cTemps4 = []
cTemps5 = []
cTemps6 = []
cTemps7 = []
cTemps8 = []
cTemps9 = []
plt.figure()
for material in data1.index:
    num_elements=data1.at[material,'number_of_elements']
    crit_temp = data1.at[material,'critical_temp']
    plt.plot(num_elements,crit_temp,'r.')
    if num_elements == 1:
        cTemps1.append(crit_temp)
    elif num_elements == 2:
        cTemps2.append(crit_temp)
    elif num_elements == 3:
        cTemps3.append(crit_temp)
    elif num_elements == 4:
        cTemps4.append(crit_temp)
    elif num_elements == 5:
        cTemps5.append(crit_temp)
    elif num_elements == 6:
        cTemps6.append(crit_temp)
    elif num_elements == 7:
        cTemps7.append(crit_temp)
    elif num_elements == 8:
        cTemps8.append(crit_temp)
    elif num_elements == 9:
        cTemps9.append(crit_temp)
cTempsavg = [np.array(cTemps1).mean(),np.array(cTemps2).mean(),np.array(cTemps3).mean(),np.array(cTemps4).mean(),np.array(cTemps5).mean(),np.array(cTemps6).mean(),np.array(cTemps7).mean(),np.array(cTemps8).mean(),np.array(cTemps9).mean()]
plt.plot([1,2,3,4,5,6,7,8,9],cTempsavg,'b*',ms=20,label='Average')
#plt.text(2,175,f'The highest critical temp is {maxTc} in the material $H_2S$')
plt.title('Number of Elements & Critical Temperatures',fontsize=10)
plt.ylabel('Critical Temp (K)',fontsize=10)
plt.xlabel('Number of Elements',fontsize=10)
plt.legend()
plt.show()
```

### Curve fitting averages

```python
plt.plot([1,2,3,4,5,6,7,8,9],cTempsavg,'b*',ms=15,label="Averages")

qfit= np.polyfit([1,2,3,4,5,6,7,8,9],cTempsavg,2,full=True)
a = qfit[0][0]
b = qfit[0][1]
c = qfit[0][2]
qerror = qfit[1][0]

x = np.linspace(0,9.5,100)
plt.plot(x,a*x**2+b*x+c,'r--',label='Quadratic Fit')
print(f'y={a:.2f}n^2+{b:.2f}n{c:.2f}')

lfit = np.polyfit([1,2,3,4,5,6,7,8,9],cTempsavg,1,full=True)
slope = lfit[0][0]
yint = lfit[0][1]
lerror = lfit[1][0]

x = np.linspace(0,9.5,100)
plt.plot(x,slope*x+yint,'g',label='Linear Fit')

plt.title('Number of Elements v Critical Temperature')
plt.text(5,20,f'Quadratic Error: {qerror:.3f}')
plt.text(5,12,f'$T_c={a:.2f}n^2+{b:.2f}n{c:.2f}$')
plt.text(5,0,f'Linear Error: {lerror:.3f}')
plt.text(5,-8,f'$T_c={slope:.3f}n{yint:.3f}$')
plt.xlabel("Number of Elements")
plt.ylabel('Critical Temperature (K)')
plt.legend()
plt.show()
```
